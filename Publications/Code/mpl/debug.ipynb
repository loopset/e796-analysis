```python
reload(u)
qw = u.DataInterface("../Events/debug_qweight.root")
nqw = u.DataInterface("../Events/debug_NO_qweight.root")

u.init_figure("delta", figsize=(8, 4))
plt.subplot(121)
mesh = qw.draw()
mesh.set_clim(vmax=1000)
qw.draw_line(0, min=0, max=128, marker="none", lw=2, color="crimson")
plt.annotate(rf"Q-weighted: $\chi^2 = ${qw.lines[0].GetChi2():.4f}",xy=(20, 100), fontsize=16)

plt.subplot(122)
mesh = nqw.draw()
mesh.set_clim(vmax=1000)
nqw.draw_line(0, min=0, max=128, marker="none", lw=2, color="crimson")
plt.annotate(rf"Not Q-weighted: $\chi^2 = ${nqw.lines[0].GetChi2():.4f}",xy=(20, 100), fontsize=16)

plt.tight_layout()
plt.show()
plt.savefig("./Outputs/debug.png")
# plt.gcf().savefig("./Outputs/delta_event.pdf")
```
